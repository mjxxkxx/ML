### Ensemble

#### 1) Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
rf = RandomForestClassifier(random_state=10, criterion = 'entropy')
```

```python
iris_data = load_iris()
```

```python
X_train , X_test , y_train , y_test = train_test_split(iris_data.data, iris_data.target,
                                                       test_size=0.3,  random_state=11)
```

```python
iris_data.target
```

```python
rf.fit(X_train, y_train)
```

```python
pred = rf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, pred)
print('랜덤 포레스트의 정확도는?: {0:.4f}'.format(accuracy))
```

```python
confusion_matrix(y_test, pred)
```

```python
ft_importance_values = rf.feature_importances_
ft_importance_values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

ft_importance_values = rf.feature_importances_
ft_importances =pd.Series(ft_importance_values, index=iris_data.feature_names)

ft_importances
```

```python
sort_ft =  ft_importances.sort_values(ascending=False)   # 내림차순 정렬

plt.figure(figsize=(10,5))    # 차트의 기본 크기 설정
plt.title('Feature Importances of Iris Dataset in Random Forest')    # 차트 제목
sns.barplot(x = sort_ft, y = sort_ft.index)    # 차트의 그래프 종류와 x축, y축 정의
plt.show()
```

```python
rf_1 =  RandomForestClassifier(
    n_estimators=30,  # 30개의 트리
    max_depth=5,      # 트리 최대 깊이 5
    random_state=1)   # 시드값 고정
```

```python
rf_1.fit(X_train, y_train)

pred_1 = rf_1.predict(X_test)
accuracy = accuracy_score(y_test, pred_1)
print('랜덤 포레스트 1번의 정확도: {0:.4f}'.format(accuracy))
```

```python
rf_2 = RandomForestClassifier(
    n_estimators = 100,   # 100개의 트리
    max_depth = 5,        # 트리 최대 깊이 5
    random_state = 1)     # 시드값 고정

rf_2.fit(X_train, y_train)

pred_2 = rf_2.predict(X_test)
accuracy = accuracy_score(y_test, pred_2)
print('랜덤 포레스트 2번의 정확도: {0:.4f}'.format(accuracy))
```

```python
rf_3 = RandomForestClassifier(
    n_estimators = 30,    # 30개의 트리
    max_depth = 1,        # 트리 최대 깊이 1
    random_state = 1)     # 시드값 고정

rf_3.fit(X_train, y_train)

pred_3 = rf_3.predict(X_test)
accuracy = accuracy_score(y_test, pred_3)
print('랜덤 포레스트 3번의 정확도는?: {0:.4f}'.format(accuracy))
```

```python
import xgboost as xgb
from xgboost import plot_importance
```

```python
iris_data = load_iris()
```

```python
X_train , X_test , y_train , y_test = train_test_split(iris_data.data, iris_data.target,
                                                       test_size=0.3,  random_state=11)
```

![대체 텍스트](https://i.postimg.cc/wTJTXSnX/3.png)

```python
xgb_model = xgb.XGBClassifier()

xgb_model.fit(X_train, y_train)
```

```python
params = { 'max_depth': 2,                      # 트리 깊이
           'learning_rate': 0.01,               # Step size
           'n_estimators': 50,                  # 트리의 갯수
           'objective':'multi:softmax',         # 목적 함수
           'num_class': 3}  # 클래스의 개수
```

```python
predict = xgb_model.predict(X_test)

print("정확도: {0:.4f}".format(accuracy_score(y_test, predict)))
```

```python
iris_data.feature_names
```

```python
xgb.plot_importance(xgb_model)
plt.show()
```

```python
xgb_model = xgb.XGBClassifier(max_depth=1, learning_rate=0.01, n_estimators=500)
xgb_model.fit(X_train, y_train)

predict = xgb_model.predict(X_test)

print("정확도: {0:.4f}".format(accuracy_score(y_test, predict)))
```

```python
for i in range(10):
  xgb.plot_tree(xgb_model, num_trees=i)
plt.show()
```
